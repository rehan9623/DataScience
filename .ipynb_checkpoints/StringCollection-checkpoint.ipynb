```python
#1.Write a program to find the length of a string without using len().
str = input("Write What you want to write")
str2 = len(str)
print(f"The length of [{str}]is {str2}")
```

```python
#2.Count the number of vowels in a string.
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in string:
    if char in vowels:
        count += 1
print(f"The number of vowels in the given string is: {count}")
```

```python
#3.Reverse a string using slicing.
str = input("Enter a String")
sub_str = str[::-1] #Ye Yaad Rakhna h Reverse k liye
print(f"The reverse of {str} is {sub_str}")
```

```python
#4.Check if a given string is a palindrome or not
str = input("Enter a string")
re_str = str[::-1]
print(f"The Reverse of {str} is {re_str}")
if str == re_str :
    print("The given String is Palindrome")
else:
    print("The given String is not Palindrome")
```

```python
#5.Find the number of occurrences of a character in a string.
str = input("Enter a String")
char2 = input("Enter a Character that you count in string :")
count = 0
for char in str:
    if char == char2 :
        count +=1
print(f"The number of occurance of {char2} in {str} is {count}")           
```

```python
#6.Replace all spaces in a string with underscores.
str = input("Enter a String")
str2 = str.replace(" ","_")
print(f"The Final String is {str2}")
```

```python
#7.Take a list of numbers and print only even numbers
numbers = [10, 15, 22, 33, 44, 55, 66]
print("Even numbers in the list are:")
for num in numbers:
    if num % 2 == 0:
        print(num)

```

```python
#8.Find the maximum and minimum numbers in a list.
l = [2,3,4,1,2,3,4,5,6,7,8,7,6,5,8,9]
max_number =max(l)
min_number = min(l)
print(f"The maximum number is {max_number}")
print(f"The minimum number is {min_number}")
```

```python
#11.Convert a tuple into a list and back into a tuple.
tuple1 = (1,2,3,4,5,6,7,11,21,11,2,9)
print(tuple1)
list1 = list(tuple1)
print(list1)
tuple2 = tuple(list1)
print(tuple2)
```

```python
#12.Write a program to merge two dictionaries.
dict1 = {
    "Name" : "Rehan",
    "Age" : 21,
    "Salary" : 45000}
print(dict1)
dict2 = {
     "Name" : "Zoya",
    "Age" : 20,
    "Salary" : 90000}
print(dict2)
dict1.update(dict2)
print(dict1)
#key valuues change kr k dict 2 k values aa jaynge     
```

```python
#13.Create a dictionary to store names and ages of 3 people, and print them.
people = {
    "Rehan": 21,
    "Zoya": 20,
    "Ayaan": 25
}
for name, age in people.items():
    print(f"{name} is {age} years old.")
```

```python
#14.Find the intersection and union of two sets
s1 = {2,12,23,21,12,33,21,21,23,12,45,65}
s2 = {4,21,23,44,54,21,22,44,67,87,11,12}
s3 = s1.union(s2)
print(s3)
s4 = s1.intersection(s2)
print(s4)
```

```python
#15.Check whether an element exists in a set or not.
# Program to check whether an element exists in a set or not

my_set = {10, 20, 30, 40, 50}

num = int(input("Enter a number to check: "))

if num in my_set:
    print(f"{num} exists in the set.")
else:
    print(f"{num} does not exist in the set.")
```
