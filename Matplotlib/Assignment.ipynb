Create a plot comparing Kohli, Rohit Sharma, and Dhawan across 10 years of runs.
Use:

- Labels
- Legends
- Colors
- Line styles
- One custom style

```python
import matplotlib.pyplot as plt
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
Virat = [1054, 623, 739, 1460, 1202, 1377, 431, 129, 302, 427]
Rohit =  [264, 815, 564, 1293, 1030, 1490, 0, 90, 249, 1255]
Dhawan = [814, 1197, 287, 864, 897, 701, 135, 297, 688, 0]
plt.figure(figsize=(10,6))
plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Virat / Rohit / Dhawan")
plt.plot(years,Virat,'r^-',label = "Virat Kohli")
plt.plot(years,Rohit,'go--',label = "Rohit Sharma")
plt.plot(years,Dhawan,'b.:',label = "Shekhar Dhawan")
plt.grid(True)
plt.legend()
plt.show()
```

## Assignment:2
Student Time Distribution Pie Chart

activities = ['Study', 'Sleep', 'Entertainment', 'Exercise', 'Other']

hours = [6, 8, 4, 2, 4]

Your pie chart must include:
- Labels for each slice
- Custom colors for all slices
- Explode the Study slice slightly
- Show percentage values on each slice
- Start the chart at 90 degrees
- Add a title to the chart
- Ensure the pie chart looks perfectly circular

Bonus (Optional)
- Add a shadow
- Change slice edge color using wedgeprops

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
activities= ["Study", "Sleep", "Entertainment", "Exercise", "Other"]
hours = [6, 8, 4, 2, 4]
colors = ["#6aaae6","#2d9683","#8a1517","#f7f414","#27e312"]
explode = [0.1,0,0,0,0]
plt.title("Student Time Distribution")
plt.axis('equal')
plt.pie(
    hours,
    labels=activities,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': '--'}
)
plt.show()
```

## Assignment:3
Weekly Study Time Stack Plot
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

math =    [2, 3, 4, 3, 2, 1, 0]
science = [1, 2, 2, 3, 2, 2, 1]
english = [1, 1, 1, 2, 1, 1, 1]
cs =      [2, 2, 3, 4, 3, 2, 1]

Your stack plot must include:
- plt.stackplot()
- Labels for each subjecs
- Custom colors
- A legend
- X-axis and Y-axis labels
- A title
- Use one matplotlib style (example: ggplot)

Bonus (Optional)

- Add alpha (transparency)
- Change legend location
- Use grid

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']

math =    [2, 3, 4, 3, 2, 1, 0]
science = [1, 2, 2, 3, 2, 2, 1]
english = [1, 1, 1, 2, 1, 1, 1]
cs =      [2, 2, 3, 4, 3, 2, 1]

labels = ['Math','Science','English','CS']
colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
plt.figure(figsize=(10,6))
plt.stackplot(days,
              math,
              science,
              english,
              cs,
              labels=labels,
              colors=colors,
              alpha=0.8)
plt.legend(loc='upper left')  
plt.title('Weekly Study Time')
plt.xlabel('Day')
plt.ylabel('Hours')
plt.grid(True)
plt.show()
```

Student Exam Marks Histogram

marks = [45, 55, 67, 72, 88, 90, 95, 78, 84, 60,
         73, 69, 58, 62, 47, 51, 66, 70, 75, 80,
         85, 92, 48, 53, 59, 64, 71, 77, 83, 89]

Your histogram must include:

- plt.hist()
- Number of bins = 5
- Custom color
- Black edge color for bars
- X-axis label and Y-axis label
- A title
- Enable grid
- Use one matplotlib style (example: ggplot)

Bonus (Optional)
- Change alpha (transparency)
- Try histtype='bar'
- Show cumulative histogram

```python
import matplotlib.pyplot as plt
plt.style.use("ggplot") 
marks = [45, 55, 67, 72, 88, 90, 95, 78, 84, 60,
         73, 69, 58, 62, 47, 51, 66, 70, 75, 80,
         85, 92, 48, 53, 59, 64, 71, 77, 83, 89]
plt.hist(marks,bins=5,edgecolor='black',color='skyblue')
plt.title("Student Exam Marks")
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
