```python
#5.Write a Python program to swap two numbers without using a third variable.
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("\nBefore swapping:")
print("a =", a)
print("b =", b)

a, b = b, a

print("\nAfter swapping:")
print("a =", a)
print("b =", b)
```

```python
#8.Demonstrate the use of all arithmetic, relational, and logical operators in Python
num1 = int(input("Enter a Number:"))
num2  = int(input("Enter a second number:"))
print("The Arithmetic Operations On the Numbers are as Follows")
print("Addition")
a = num1+num2
print(a)
print("Subtraction")
b = num1-num2
print(b)
print("Multiplication")
c = a*b
print(c)
print("Division")
d = a/b
print(d)
print("Floor Divisio")
e = a//b
print(e)
print("Modulus")
f = a%b
print(f)
print("Exponential")
g = a**b
print(g)
print("The Relational Operations On the Number are as follows")
print(f"{a} == {b} :", a == b)   
print(f"{a} != {b} :", a != b)  
print(f"{a} > {b}  :", a > b)  
print(f"{a} < {b}  :", a < b)    
print(f"{a} >= {b} :", a >= b)   
print(f"{a} <= {b} :", a <= b)   
print("The Logical Operations On the Number are as follows")
print("Both conditions true (a > 0 and b > 0):", a > 0 and b > 0)
print("At least one condition true (a > 0 or b > 0):", a > 0 or b > 0)
print("Reversing condition (not(a > b)):", not(a > b))
```
