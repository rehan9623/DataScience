```python
#6.Write a program to find factorial of a number using a for loop.
num = int(input("Enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print("Factorial =", factorial)    
```

```python
#8.Write a program to check if a number is prime or not.
num = int(input("Enter a number:"))
if num <= 1:
    print("Not a prime number")
else:
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

```python
#9.Print numbers from 1–200 but stop when the sum of printed numbers first exceeds 500.
count = 0
for i in range(1, 201):
    count += i
    print(i)
    if count > 500:
        break
print("Sum exceeded 500 at number:", i)
```

```python
#12.Write a python program to print vowel of given string
str1 = input("Enter a string: ")
vowels = "aeiouAEIOU"
found = []
for ch in str1:
    if ch in vowels:
        found.append(ch)
print("Vowels found:", found)
print("Total vowels:", len(found))
```

```python
#14.write a python program to print number of digits in given number
num = input("Enter a number: ")
count = 0
for ch in num:
    if ch.isdigit():   # Count only digit characters
        count += 1
print("Number of digits =", count)
```

```python
#15.#14.write a python program to print number of digits in given number
num = input("Enter a number: ")
count = 0
for ch in num:
    if ch >= '0' and ch <= '9':
        count += 1
print("Number of digits =", count)
```
