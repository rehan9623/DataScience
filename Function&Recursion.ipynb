```python
#5.Write a recursive function to print Fibonacci sequence up to n terms.
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)
terms = int(input("Enter number of terms: "))
print("Fibonacci sequence:")
for i in range(terms):
    print(fibonacci(i), end=" ")
```

```python
#9.Write a python function to calculate compund intrest(TSRS)
def compound(p, r, n, t):
    amount = p * (1 + (r / (100 * n))) ** (n * t)
    ci = amount - p
    return ci
p = float(input("Enter principal amount: "))
r = float(input("Enter annual interest rate: "))
n = float(input("Enter number of compounding periods per year: "))
t = float(input("Enter total years: "))
result = compound(p, r, n, t)
print("Compound Interest =", result)
```

```python
#13.write python function to check wheater a number is prime or not.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number: "))
result = is_prime(num)

if result:
    print(num, "is a Prime Number")
else:
    print(num, "is NOT a Prime Number")
```

```python
#14.write python function check if an year is leap year (TSRS)
def is_leap(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False

yr = int(input("Enter a year: "))
result = is_leap(yr)

if result:
    print(yr, "is a Leap Year")
else:
    print(yr, "is NOT a Leap Year")
```

```python
#19.write a python function to calcultae lCM of two number(TSRS)
def lcm(a, b):
    greater = max(a, b)
    while True:
        if (greater % a == 0) and (greater % b == 0):
            return greater
        greater += 1
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
result = lcm(n1, n2)
print("LCM of", n1, "and", n2, "is:", result)
```

```python
#20.write a python function to create a  list of prime number in between two given number(TSRS)
def prime_list(a, b):
    primes = []
    for num in range(a, b + 1):
        if num > 1:
            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    break
            else:
                primes.append(num)
    return primes
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))
result = prime_list(start, end)
print("Prime numbers between", start, "and", end, "are:", result)
```

```python
#21.write a python function to remove duplicate elements from a given list(TSRS)
def remove_duplicates(lst):
    new_list = []
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    return new_list
elements = list(map(int, input("Enter list elements separated by space: ").split()))
result = remove_duplicates(elements)
print("List after removing duplicates:", result)
```

```python
#22.write a python function to find numbers in a given text,store number in list and return list.(TSRS)
def extract_numbers(text):
    num = ""
    numbers = []
    
    for ch in text:
        if ch.isdigit():
            num += ch
        else:
            if num != "":
                numbers.append(int(num))
                num = ""
    
    # Add last number if text ends with a digit
    if num != "":
        numbers.append(int(num))
    
    return numbers

# Example
text = "I have 2 apples, 35 bananas and 1000 rupees."
print(extract_numbers(text))
```

```python
#38.write a recursive function to calculate sum of digit of a given number.
def sum_of_digits(n):
    n = abs(n)  #
    if n == 0:
        return 0
    else:
        return (n % 10) + sum_of_digits(n // 10)
num = int(input("Enter a number: "))
print("Sum of digits =", sum_of_digits(num))
```
